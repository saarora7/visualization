```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613) #Setting up the random seed for reproducability
x = np.arange(50)
y = np.random.randint(0, 100,50)  #  In Y axis generate 50 numbers randomly between 1 to 100

fig, ax = plt.subplots(figsize=(5, 3)) # Generate a figure , it should a plot of size 5 by 3
ax.scatter(x,y) # generate Scatter Plot 
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

ax.set_title('Total growth over time') # Adding Title to the Figure
ax.set_ylabel('Total growth') # Adding Title to Y axis
ax.set_xlabel('Years since start') # Adding Title to X axis
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20} # Creating a font with family named Sans-serif, colour is blue and size is 20
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc = 'left') # Loc = 'let' shifts the Title to the left and placing font1 generates title font with colour blue and size 20
ax.set_ylabel('Total growth', fontdict = font2) # Font 2 sets the font of Y axis lable to color green and size 24
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y, marker='*', color = "indigo") # marker gives a specific marker to the points in plot. We can specify them by replacing * with +,D,> etc and color is indigo

# Can change marker with +,>,D etc
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y, 
    marker='*',
    markersize = 12, # Changinh the size of marker
    color = "indigo",
    linestyle = '--', # When the markers are connected, specifying here to connect the markers by --
    linewidth = 2, # width of the line connecting the markers is 2
    markeredgecolor = '#fa9359', # giving colour to the edges of the marker
    markerfacecolor = '#000000')  # giving colors of inside the marker

ax.grid(axis = 'y')
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y, 
    marker='*',
    markersize = 12,
    color = "indigo",
    linestyle = '--',
    linewidth = 2,
    markeredgecolor = '#fa9359',
    markerfacecolor = '#000000') 

ax.grid(axis = 'y') # adding grid inY axis
ax.grid(axis = 'y', color = "red", linewidth = 2, linestyle = '-') # Colour of gird should be red, with with of the line as 2 and line is represented as -

ax.grid(axis = 'x')
ax.grid(axis = 'x', color = "yellow", linewidth = 2, linestyle = '-')
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
values = [23, 17, 35, 29, 12]

# Create bar chart
plt.figure(figsize=(6,4)) # Creating a figure with size 6 by 4
plt.bar(categories, values, color='skyblue') # Bar graph with colour sky blue
plt.title("Fruit Quantities") # Title of figure is quantities
plt.xlabel("Fruit") # labelling X axis as Fruits
plt.ylabel("Quantity") # Labeliling Y axis as Quantity
plt.show()

```
