### 1. Software Used
The visualization was created using **Microsoft Excel**. Excel was chosen because the dataset is primarily categorical, and a bar chart could be easily generated using a PivotTable and PivotChart. The chart uses a **black background with blue bars** for a visually striking effect.

### 2. Intended Audience
- Public health officials and healthcare administrators in Toronto.  
- Policy makers and planners interested in outbreak management.  
- Researchers studying healthcare outbreaks in Ontario.

### 3. Message / Information Conveyed
This bar chart shows the **number of outbreaks across different healthcare institution settings** in Toronto. The design highlights which settings, such as **acute care hospitals, chronic care, and long-term care homes**, experience more frequent outbreaks. The chart helps prioritize outbreak monitoring and intervention.

### 4. Design Considerations
- **Background:** Black, for high contrast with blue bars.  
- **Bar Color:** Blue, to stand out clearly against the dark background.  
- **Chart Type:** Bar chart, effective for categorical comparison.  
- **Axis Labels:** X-axis shows outbreak settings; Y-axis shows number of outbreaks.  
- **Title:** Clear and descriptive to communicate the key message.  
- **Readability:** Simple layout with clear spacing to enhance visibility.  

### 5. Reproducibility
- Original CSV file is included for replication.  
- PivotTable and PivotChart steps in Excel are documented and can be repeated.  
- If Excel is unavailable, the dataset allows replication in alternative software like Python, Tableau, or Google Sheets.

### 6. Accessibility
- High-contrast color scheme (blue bars on black background).  
- Clear and legible axis labels and title.  
- Minimal clutter to reduce visual noise.

### 7. Individuals / Communities Impacted
- Toronto residents and patients who may be affected by outbreaks.  
- Healthcare staff and administrators responsible for outbreak prevention.  
- Public health researchers and planners.

### 8. Feature Selection
- Included: **Outbreak Setting** (categorical) and **_id** (count of outbreaks).  
- Excluded: Other details like Causative Agent or exact dates, to maintain a clear focus on the number of outbreaks per setting.

### 9. Underwater Labour
- Cleaning and preparing the dataset.  
- Creating the PivotTable and PivotChart with correct aggregation.  
- Designing the chart with high-contrast colors, readable labels, and a descriptive title.  
- Documenting steps to ensure reproducibility and accessibility.

**Conclusion:**  
The visualization provides a visually striking view of outbreak frequency across healthcare settings in Toronto, designed for clarity, accessibility, and immediate comprehension by technical and non-technical audiences alike.
