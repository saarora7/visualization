The visualization shows monthly trends of ferry ticket counts to Toronto Island. It highlights peak and low usage months, helping stakeholders understand seasonal demand patterns.

** Design considerations: **
Clarity: Used a line plot to clearly show trends over time.
Color: Indigo line for high contrast against the white grid background.
Labels: Clearly labeled axes and rotated x-axis ticks for readability.
Size: Large figure (12x6) to ensure easy viewing of details.
Grid: Whitegrid style to make trends easier to follow.
